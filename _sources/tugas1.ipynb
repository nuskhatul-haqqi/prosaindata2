```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
from textblob import TextBlob
import re
def clean_tweet(tweet):
  return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\s+)"," ",tweet).split())
# sentimens analisis dengan polarity menggunakan library textblob
def analize_sentiment(tweet):
  analysis = TextBlob(clean_tweet(tweet))
  if analysis.sentiment.polarity > 0:
    return 1
  elif analysis.sentiment.polarity == 0:
    return 0
  else:
    return -1
  def analize_subjectorobject(tweet):
    #untuk nilai -1 termasuk negatif, 0 netral, 1 positif
    analysis = TextBlob(clean_tweet(tweet))
    if analysis.sentiment.polarity > 0:
      return 'positif'
    elif analysis.sentiment.polarity == 0:
      return 'netral'
    else:
      return 'negatif'
```

```python
#creating list to append tweet data to
tweets_list2 = []
```

```python
#using twitterSearchScraper to scrape data and append tweets to list
for i, tweet in enumerate(sntwitter.TwitterSearchScraper('Prabowo since:2022-01-01 until:2023-02-01').get_items()):
  if i>500:
    break
  tweets_list2.append([tweet.user.username,tweet.content,analize_sentiment(tweet.content)])
```

```python
tweet2 = pd.DataFrame(tweets_list2, columns=['Username','Text','Label'])
```

## Scraping Berita Online pada Situs detik.com

```python
import requests as req
from bs4 import BeautifulSoup as bs
from datetime import datetime
import pandas as pd
hades = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
```

```python
berita_list = []
```

```python
def scrape_detik(hal):
    global hades
    a = 1
    for page in range(1,hal):
        url = f'https://www.detik.com/search/searchnews?query=pemilu+2024&sortby=time&page={page}'
        ge = req.get(url,hades).text
        sop = bs(ge,'lxml')
        li = sop.find('div',class_='list media_rows list-berita')
        lin = li.find_all('article')
        for x in lin:
            link = x.find('a')['href']
            date = x.find('a').find('span',class_='date').text.replace('WIB','').replace('detikNews','').split(',')[1]
            headline = x.find('a').find('h2').text
            ge_ = req.get(link,hades).text
            sop_ = bs(ge_,'lxml')
            content = sop_.find_all('div',class_='detail__body-text itp_bodycontent')
            for x in content:
                x = x.find_all('p')
                y  = [y.text for y in x ]
                content_ = ''.join(y).replace('\n', '').replace('ADVERTISEMENT','').replace('SCROLL TO RESUME CONTENT','')
                a += 1
                berita_list.append([headline,content_])
    print(f'selesai dengan [{a}] data')
```

```python
scrape_detik(7)
```

```python
berita = pd.DataFrame(berita_list, columns=['Judul','Isi'])
```

## Scraping PTA Teknik Informatika

```python
import requests as req
from bs4 import BeautifulSoup as bs
from datetime import datetime
import pandas as pd
hades = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
```

```python
pta_list = []
```

```python
from typing_extensions import LiteralString
def scrape_pta(hal):
    global hades
    a = 1
    for page in range(1,hal):
        url = f'https://pta.trunojoyo.ac.id/c_search/byprod/10/{page}'
        ge = req.get(url,hades).text
        sop = bs(ge,'lxml')
        lii = sop.find('ul',class_='items list_style') 
        lin = lii.find_all('li')
        for x in lin:
            link = x.find('a', class_='gray button')['href']
            headline = x.find('div').find('a',class_='title').text.replace('\r\n','')
            penulis = x.find('div').find('span').text.replace('Penulis :','')
            ge_ = req.get(link,hades).text
            sop_ = bs(ge_,'lxml')
            content = sop_.find_all('ul',class_='items list_style')
            for x in content:
                content_ = x.find('p').text
                print(f'done[{a}] > {headline[0:10]}')
                a += 1
                pta_list.append([headline,penulis,content_])
    print(f'selesai dengan [{a}] data')
```

```python
scrape_pta(173)
```

```python
pta_ = pd.DataFrame(pta_list, columns=['Judul','Penulis','Isi'])
```
